```python
import json
import requests
import time
def getlnglat(address):
#     key = 'DBEBZ-GHJ35-2PAIB-QNWLF-W6TI3-6EBXQ'
#     parameters = {'address': address, 'key': key}
#     base = "https://apis.map.qq.com/ws/geocoder/v1/?"
    key = 'iz8HxdIV4IiiWOYcA1t01qdUMeVUCPEi'
    parameters = {'address': address, 'ak': key, 'output': 'json'}
    base = 'http://api.map.baidu.com/geocoder/v2/?'
    response = requests.get(base, parameters)
    answer = response.json()
    time.sleep(0.005) # 1s per 200 records
    if answer['status'] == 0:
        return answer['result'], answer['status'], 'success'
    return None, answer['status'], answer['msg']
```

```python
output = []
start = len(df.index)
end = len(data.index)
print('Processing data from index %s to %s' % (start, end))
for indexs in data.index[start:end]:
    location = data.loc[indexs, 'Address']
    geo = {'queryLoc': location}
    result, status, msg = getlnglat(location)
    geo['answer'] = result
    geo['status'] = status
    geo['message'] = msg
    output.append(geo)
```

```python
from pandas.io.json import json_normalize
outputDf = json_normalize(output)
# status: 
# 0 = query ok
# 1 = no result
# 2 = address too long
outputDf
```

```python
df = df.append(outputDf)
df = df.reset_index()
del df['index']
df
```

```python
# Be cautious, the following operation will overwrite the output.csv
df.to_csv('output.csv', encoding='utf_8_sig')
```
